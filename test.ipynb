```python
from model import GPT, GPTConfig
import torch

# Configuration d'un mini modèle GPT (très petit pour le test)
config = GPTConfig(
    vocab_size=50,    # par ex. 50 tokens distincts 
    block_size=8,     # contexte de 8 tokens
    n_layer=2, 
    n_head=4, 
    n_embd=64
)
model = GPT(config)

# Générer une séquence "dummy" de 20 tokens à partir d'une entrée aléatoire de 8 tokens :
idx = torch.randint(0, 50, (1, 8), dtype=torch.long)  # séquence aléatoire de taille 8
generated_idx = model.generate(idx, max_new_tokens=20)  # génération de 20 tokens supplémentaires
print(generated_idx)
```

```python
from datasets import load_dataset

dataset = load_dataset("wikitext", "wikitext-2-raw-v1")

def save_split(split, filename):
    with open(filename, "w", encoding="utf-8") as f:
        for line in dataset[split]["text"]:
            if line.strip():
                f.write(line + "\n")

save_split("train", "train.txt")
save_split("validation", "val.txt")
save_split("test", "test.txt")
```
